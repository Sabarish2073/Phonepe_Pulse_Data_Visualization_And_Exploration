```python
import os
import json
import pandas as pd
import psycopg2
```

```python
#aggr_insurance
path1="C:/Users/ASUS 502/OneDrive/Desktop/phonepe data/pulse/data/aggregated/insurance/country/india/state/"
aggr_insurance_list=os.listdir(path1)

columns1={"State":[],"Year":[],"Quarter":[],"Transaction_type":[],"Transaction_count":[],"Transaction_amount":[]}
for state in aggr_insurance_list:
    cur_state=path1 + state + "/"
    aggr_year_list=os.listdir(cur_state)
    for year in aggr_year_list:
        cur_year=cur_state + year + "/"
        aggr_file_list=os.listdir(cur_year)
        for file in aggr_file_list:
            cur_file=cur_year + file
            
            data=open(cur_file,"r")
            A=json.load(data)
            #print(A)

            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns1["Transaction_type"].append(name)
                columns1["Transaction_count"].append(count)
                columns1["Transaction_amount"].append(amount)
                columns1["State"].append(state)
                columns1["Year"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
aggr_insurance=pd.DataFrame(columns1)
aggr_insurance["State"]=aggr_insurance["State"].str.replace("andaman-&-nicobar-islands","andaman-&-nicobar")
aggr_insurance["State"]=aggr_insurance["State"].str.replace("-"," ")
aggr_insurance["State"]=aggr_insurance["State"].str.title()
aggr_insurance["State"]=aggr_insurance["State"].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")


```

```python
#aggr_transaction
path2="C:/Users/ASUS 502/OneDrive/Desktop/phonepe data/pulse/data/aggregated/transaction/country/india/state/"
aggre_tran_list=os.listdir(path2)
columns2={"State":[],"Year":[],"Quarter":[],"Transaction_type":[],"Transaction_count":[],"Transaction_amount":[]}
for state in aggre_tran_list:
    cur_state=path2 + state +"/"
    #print(cur_state)
    aggr_year_list=os.listdir(cur_state)
    #print(aggr_year_list)
    for year in aggr_year_list:
        cur_year=cur_state + year + "/"
        #print(cur_year)
        aggr_file_list=os.listdir(cur_year)
        #print(aggr_file_list)
        for file in aggr_file_list:
            cur_file=cur_year + file
            data=open(cur_file,"r")
            #print(data)
            B=json.load(data)
            for i in B["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns2["Transaction_type"].append(name)
                columns2["Transaction_count"].append(count)
                columns2["Transaction_amount"].append(amount)
                columns2["State"].append(state)
                columns2["Year"].append(year)
                columns2["Quarter"].append(int(file.strip(".json")))
                #print(columns1)
                
aggr_transaction=pd.DataFrame(columns2)
aggr_transaction["State"]=aggr_transaction["State"].str.replace("andaman-&-nicobar-islands","andaman-&-nicobar")
aggr_transaction["State"]=aggr_transaction["State"].str.replace("-"," ")
aggr_transaction["State"]=aggr_transaction["State"].str.title()
aggr_transaction["State"]=aggr_transaction["State"].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")

                
```

```python
#aggr_user

path3="C:/Users/ASUS 502/OneDrive/Desktop/phonepe data/pulse/data/aggregated/user/country/india/state/"
aggr_user_list=os.listdir(path3)
#print(aggr_user)
columns3={"State":[],"Year":[],"Quarter":[],"Brands":[],"Count":[],"Percentage":[]}
for state in aggr_user_list:
    cur_state=path3 + state + "/"
    aggr_year_list=os.listdir(cur_state)
    for year in aggr_year_list:
        cur_year=cur_state + year + "/"
        aggr_file_list=os.listdir(cur_year)
        for file in aggr_file_list:
            cur_file=cur_year + file
            
            data=open(cur_file,"r")
            C=json.load(data)
            #print(B)
            try:
                for i in C["data"]["usersByDevice"]:
                    brand=i["brand"]
                    count=i["count"]
                    percentage=i["percentage"]
                    columns3["Brands"].append(brand)
                    columns3["Count"].append(count)
                    columns3["Percentage"].append(percentage)
                    columns3["State"].append(state)
                    columns3["Year"].append(year)
                    columns3["Quarter"].append(int(file.strip(".json")))
               
            except:
                pass
            
aggr_user=pd.DataFrame(columns3)
aggr_user["State"]=aggr_user["State"].str.replace("andaman-&-nicobar-islands","andaman-&-nicobar")
aggr_user["State"]=aggr_user["State"].str.replace("-"," ")
aggr_user["State"]=aggr_user["State"].str.title()
aggr_user["State"]=aggr_user["State"].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")
```
